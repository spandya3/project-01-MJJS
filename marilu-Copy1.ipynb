```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df= pd.read_csv("clean_movie_data.csv")
df.head()
```

```python
# check the Genre
df["Genre"].unique()
```

```python
df.columns
```

```python
# Average movie duration
df['Duration (min)'].mean()
```

```python
# Average duration by Genre
duration_by_genre= df[["Genre",'Duration (min)']].groupby(["Genre"]).mean().sort_values(by="Duration (min)", ascending=False)
```

```python
 duration_by_genre
```

```python
duration_by_genre.head(30).plot(kind="bar",
                                 ylabel="Average Duration (minutes)",
                                 title= "Average duration by genre.(top 20)",
                                figsize=(15,8), color='r') 
plt.show()
```

```python
df= pd.read_csv("clean_movie_data.csv")
df.head()


```

```python
df.info()
```

```python
df["Votes"].str.replace(",","").astype(int)
```

```python
df["Votes"]= df["Votes"].str.replace(",","").astype(int)
voting_df =df[['Title','Votes','Year']]
sorted_voting_df= voting_df.sort_values('Votes', ascending=False)
voting_df_2023= sorted_voting_df[sorted_voting_df["Year"]==2023.0]
voting_df_2023 
```

```python
voting_df_2023.head().set_index("Title").plot(kind="bar",y="Votes",
                                 ylabel="Votes",
                                 title= "Top 5 movies by votes in 2023",
                                figsize=(15,12), color='gray') 
plt.show()

```
