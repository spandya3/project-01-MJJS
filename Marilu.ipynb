```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df= pd.read_csv("clean_movie_data.csv")
df.head()
```

```python
# check the Genre
df["Genre"].unique()
```

```python
df.columns
```

```python
# Average movie duration
df['Duration (min)'].mean()
```

```python
# Average duration by Genre
duration_by_genre= df[["Genre",'Duration (min)']].groupby(["Genre"]).mean().sort_values(by="Duration (min)", ascending=False)
```

```python
 duration_by_genre
```

```python
duration_by_genre.head(30).plot(kind="bar",
                                 ylabel="Average Duration (minutes)",
                                 title= "Average duration by genre.(top 20)",
                                figsize=(15,12), color='r') 
plt.show()
```
